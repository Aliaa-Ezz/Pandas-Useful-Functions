```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
df = pd.read_csv(url)
```

```python
df_EG = df[df['Country/Region'] == 'Egypt']
df_EG
```

```python
del df_EG['Province/State']
del df_EG['Lat']
del df_EG['Long']
```

```python
df_EG_Clns = pd.DataFrame(df_EG.unstack()).reset_index()
df_EG_Clns
```

```python
del df_EG_Clns['level_1']
```

```python
df_EG_Clns = df_EG_Clns.iloc[1:]
df_EG_Clns
```

```python
df_EG_Clns = df_EG_Clns.rename(columns = {'level_0':'Date', 0:'cumulative'})
df_EG_Clns
```

```python
df_EG_Clns['cumulative'] = df_EG_Clns['cumulative'].diff()
df_EG_Clns
```
